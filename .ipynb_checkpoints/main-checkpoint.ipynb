```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import utils
from tensorflow.keras.applications import vgg16, resnet50
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Dropout, Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import cv2
```

# 1.Load Data

```python
directory = './brain_tumor_dataset'
image_size = (256, 256, 3)
```

```python
train_data = utils.image_dataset_from_directory(
    directory,
    seed=100,
    validation_split=0.2,
    subset='training',
    image_size=image_size[:2],
)
```

```python
val_data = utils.image_dataset_from_directory(
    directory,
    seed=100,
    validation_split=0.2,
    subset='validation',
    image_size=image_size[:2],
)
```

# 2.Visualize Data

```python
plt.figure(figsize=(15, 15))
images = train_data.take(1).as_numpy_iterator().next()[0]
labels = train_data.take(1).as_numpy_iterator().next()[1]
for i in range(10):
    img = images[i].astype('uint8')
    label = train_data.class_names[labels[i]]
    ax = plt.subplot(5, 5, i+1)
    plt.imshow(img)
    plt.title(label)
    plt.axis('off')
```

# 3.Normalization

```python
norm_train_data = train_data.map(lambda x, y : (x/255, y))
norm_val_data = val_data.map(lambda x, y : (x/255, y))
```

# 4.Build Model

```python
model = tf.keras.models.Sequential()
```

```python
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=image_size))
model.add(MaxPooling2D())
model.add(Conv2D(316, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

# 5.Train

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', 'Recall'])
```

```python
history = model.fit(
    train_data,
    epochs=10,
    validation_data=val_data,
    callbacks=tf.keras.callbacks.TensorBoard(),
)
```

### Plot Performance

```python
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
```

```python
plt.plot(train_accuracy, label='train')
plt.plot(val_accuracy, label='validation')
plt.title('accuracy')
plt.legend()
plt.show()
```
